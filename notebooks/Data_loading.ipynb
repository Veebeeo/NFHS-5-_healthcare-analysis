```python
import pandas as pd
import numpy as np
import os
import json
from pathlib import Path

PROJECT_ROOT = Path("..").resolve()
RAW_DIR = PROJECT_ROOT / "data" / "raw"
GEO_DIR = PROJECT_ROOT / "data" / "geospatial"

print(f"Project root: {PROJECT_ROOT}")
print(f"Raw data dir: {RAW_DIR}")
print(f"Geo data dir: {GEO_DIR}")
print(f"\nFiles in raw dir: {os.listdir(RAW_DIR)}")
print(f"Files in geo dir: {os.listdir(GEO_DIR)}")
```

```python
#loading the dataset
df_india = pd.read_csv(RAW_DIR / "India.csv", encoding='utf-8-sig')
print("=" * 60)
print("INDIA.CSV : MAIN NFHS-5 DATASET")
print("=" * 60)
print(f"Shape: {df_india.shape}")
print(f"Columns: {list(df_india.columns)}")
print(f"\nFirst 5 rows:")
df_india.head()
```

```python
# Comparison dataset (nfhs5 vs nfhs4)
df_change = pd.read_csv(RAW_DIR / "India_Change.csv", encoding='utf-8-sig')

print("=" * 60)
print("INDIA_CHANGE.CSV : NFHS-5 vs NFHS-4 COMPARISON")
print("=" * 60)
print(f"Shape: {df_change.shape}")
print(f"Columns: {list(df_change.columns)}")
print(f"\nData types:\n{df_change.dtypes}")
print(f"\nFirst 5 rows:")
df_change.head()
```

```python
print("KEY STATISTICS OF INDIA_CHANGE.CSV")
print(f"\n Total rows: {len(df_change):,}")
print(f" Unique states/UTs: {df_change['State'].nunique()}")
print(f" Unique districts: {df_change['District Name'].nunique()}")
print(f" Unique categories: {df_change['Category'].nunique()}")
print(f" Unique indicators: {df_change['Indicator'].nunique()}")

print(f"\n States/UTs list ({df_change['State'].nunique()}):")
for i, state in enumerate(sorted(df_change['State'].unique()), 1):
    print(f"   {i:2d}. {state}")

print(f"\n Indicator Categories ({df_change['Category'].nunique()}):")
for i, cat in enumerate(sorted(df_change['Category'].unique()), 1):
    print(f"   {i:2d}. {cat}")
```

```python
print("DATA QUALITY CHECK")
# missing values
print(f"\n Missing values per column:")
print(df_change.isnull().sum())

# Check non-numeric values
print(f"\nNFHS 5 column dtype: {df_change['NFHS 5'].dtype}")

# Check string values in numeric columns
if df_change['NFHS 5'].dtype == 'object':
    non_numeric = df_change[pd.to_numeric(df_change['NFHS 5'], errors='coerce').isna() & df_change['NFHS 5'].notna()]
    print(f"Non-numeric values in 'NFHS 5': {len(non_numeric)}")
    if len(non_numeric) > 0:
        print(f"   Examples: {non_numeric['NFHS 5'].unique()[:10]}")
else:
    print("NFHS 5 column is already numeric")

if df_change['NFHS 4'].dtype == 'object':
    non_numeric = df_change[pd.to_numeric(df_change['NFHS 4'], errors='coerce').isna() & df_change['NFHS 4'].notna()]
    print(f"Non-numeric values in 'NFHS 4': {len(non_numeric)}")
    if len(non_numeric) > 0:
        print(f"   Examples: {non_numeric['NFHS 4'].unique()[:10]}")
else:
    print("NFHS 4 column is already numeric")

# districts per state
print(f"\n Districts per State:")
state_dist_counts = df_change.groupby('State')['District Name'].nunique().sort_values(ascending=False)
for state, count in state_dist_counts.items():
    print(f"   {state}: {count} districts")
```

```python
print("GEOJSON VERIFICATION")
geojson_path = GEO_DIR / "Districts.geojson"
file_size_mb = os.path.getsize(geojson_path) / (1024 * 1024)
print(f"File size: {file_size_mb:.1f} MB")

with open(geojson_path, 'r', encoding='utf-8') as f:
    geo_data = json.load(f)

print(f"Type: {geo_data['type']}")
print(f"Number of features (districts): {len(geo_data['features'])}")

first_feature = geo_data['features'][0]
print(f"Properties in each feature: {list(first_feature['properties'].keys())}")
print(f"First district: {first_feature['properties']}")
```

```python
from sqlalchemy import create_engine, text
from dotenv import load_dotenv

# Load credentials from .env file
env_path = PROJECT_ROOT / ".env"
load_dotenv(env_path)

DB_HOST = os.getenv("DB_HOST", "localhost")
DB_PORT = os.getenv("DB_PORT", "5432")
DB_NAME = os.getenv("DB_NAME", "nfhs_health_analytics")
DB_USER = os.getenv("DB_USER", "postgres")
DB_PASSWORD = os.getenv("DB_PASSWORD", "")
DB_SCHEMA = os.getenv("DB_SCHEMA", "nfhs")

conn_string = f"postgresql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"

try:
    engine = create_engine(conn_string)
    with engine.connect() as conn:
        result = conn.execute(text("SELECT current_database(), current_schema(), version();"))
        row = result.fetchone()
        print("PostgreSQL CONNECTION SUCCESSFUL")
        print(f"   Database: {row[0]}")
        print(f"   Schema: {row[1]}")
        print(f"   Version: {row[2][:50]}...")

        # Check nfhs schema exists
        result = conn.execute(text(
            "SELECT schema_name FROM information_schema.schemata WHERE schema_name = 'nfhs'"
        ))
        schema_row = result.fetchone()
        if schema_row:
            print(f"Schema 'nfhs' exists")
        else:
            print(f"Schema 'nfhs' NOT FOUND")

except Exception as e:
    print(f"CONNECTION FAILED: {e}")
    print(f"\nTroubleshooting:")
    print(f"  1. Is PostgreSQL running? Check Services or pgAdmin")
    print(f"  2. Is the password correct in .env file?")
    print(f"  3. Did you create the database 'nfhs_health_analytics'?")
    print(f"  4. Connection string: {conn_string.replace(DB_PASSWORD, '****')}")
```
